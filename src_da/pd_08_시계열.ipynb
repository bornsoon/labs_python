```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'visitor': np.random.randint(1, 100, size=50)}, index=pd.date_range('2024-01-01', periods=50, freq='30min'))
df
```

```python
df['visitor'].resample('10min')
```

```python
df['visitor'].resample('10min').sum()
```

```python
df['visitor'].resample('10min').ffill()
```

```python
df['visitor'].resample('1h').sum()
```

```python
df['visitor'].resample('1h', closed='right').sum() # default: left
```

```python
# shift
df = pd.DataFrame({'visitor': np.random.randint(1, 100, size=10)}, index=pd.date_range('2024-01-01', periods=10, freq='D'))
df
```

```python
df.shift(1, fill_value=0)
```

```python
df['전일'] = df.shift(1, fill_value=0)
df
```

```python
df['증가량'] = df['visitor']-  df['전일']
df
```

```python
df['증가율'] = ((df['증가량']/df['전일'])*100).round(1)
df
```

```python
# rolling
df = pd.DataFrame({'A': np.arange(5)}, index=pd.date_range('2024-01-01', periods=5, freq='D'))
df
```

```python
df.rolling(2).mean()
```

```python
df.rolling(3).mean()
```
