```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1, 2], 'B':[3, 4]})
df2 = pd.DataFrame({'C':[5, 7], 'D':[7, 8]})
df3 = pd.DataFrame({'A':[9, 10], 'C':[11, 12], 'D':[13, 14]}, index=[1, 2])
df4 = pd.DataFrame({'B':[3, 4, 5], 'E':[15, 16, 17]})

display(df1, df2, df3, df4)
```

## append
### 열의 갯수가 같고, 시계열 데이터 업데이트할 때 자주 씀

```python
# df = df1.append(df2) 행추가
df = pd.concat([df1, df2])
df
```

```python
# df = df1.append(df2) 행추가
df = pd.concat([df1, df2], ignore_index=True)
df
```

## join
### 열 합치기, 같은 열명이 있으면 오류 발생

```python
df = df1.join(df2)  # 열 합치기, 같은 열명이 있으면 오류 발생
df
```

## concat
### 행의 갯수가 같고 열을 추가하고 싶을 때 많이 씀

```python
df = pd.concat([df1, df2], axis=1)
df = pd.concat([df1, df2], axis=1, ignore_index=True)
df
```

```python
df = pd.concat([df1, df3]) # axis=0, 열 인덱스 기준
df
```

```python
df = pd.concat([df1, df3], axis=1) # axis=1, 행 인덱스 기준
df
```

```python
df = pd.concat([df1, df3], axis=1, join='outer') # outer가 기본
display(df)

df = pd.concat([df1, df3], axis=1, join='inner') # inner가 기본
display(df)

df = pd.concat([df1, df3], axis=0, join='inner')
display(df)
```

```python
df = pd.concat([df1, df4])
df
```

```python
df = pd.concat([df1, df4], axis=1)
df
```

```python
df = df1.merge(df4)  # B컬럼을 기준으로 df1(left)가 inner조인
df
```

```python
df = df1.merge(df4, how='left', on='B')
display(df)
```

```python
df = df1.merge(df4, how='right', on='B')
display(df)
```

```python
df = df1.merge(df4, how='inner', on='B')
display(df)
```
