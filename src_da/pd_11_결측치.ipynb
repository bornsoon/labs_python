```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1, 2, 3, np.nan],
                   'B':[5, 6, 7, np.nan],
                   'C':[9, np.nan, 11, np.nan],
                   'D':[np.nan, 14, 15, np.nan]},
                   index=pd.date_range(start='2019-01-01', periods=4))
df
```

```python
df.isnull().sum() / len(df)
```

```python
df1 = df.dropna(how='all')
df1
```

```python
df1 = df.dropna(axis=1, how='all')
df1
```

```python
df1 = df.dropna(thresh=3)
df1
```

```python
df1 = df.dropna(thresh=3, axis=1)
df1
```

```python
df1 = df.dropna(subset=['B', 'C'])
df1
```

```python
df = pd.DataFrame({'A':[1, 2, 3, np.nan],
                    'B':[5, 6, 7, np.nan],
                    'C':[9, np.nan, 11, np.nan],
                    'D':[np.nan, 14, 15, np.nan]}
                    ,index=range(4), dtype='Int64') 
df
```

```python
df.isnull().sum() / len(df)
```

```python
df1 = df.fillna(0)
df1
```

```python
df1 = df.fillna({'A':0, 'B':1, 'C':2, 'D':3})
df1
```

```python
df1 = df.ffill()
df1
df2 = df.fillna(method='ffill')
df2
```

```python
df1 = df.fillna(method='ffill', axis=1)
df1
```

```python
df1 = df.bfill()
df2 = df.fillna(method='bfill')
display(df1, df2)
```

```python
df1 = df.fillna(df.mean().astype('int'))
df1
```

```python
df1 = df.interpolate()
df1
```
