```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='gulim')
```

```python
df = pd.read_csv('datasets/seaborn-data/tips.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
df = sns.load_dataset('fmri')  # github  
sns.lineplot(x='timepoint', y='signal', data=df, marker='o', linestyle='-', color='g', alpha=0.5)
plt.show()
```

```python
sns.lineplot(x='timepoint', y='signal', hue='event', style='event', data=df, markers=True, dashes=False, errorbar=None, alpha=0.8)
plt.show()
```

```python
df = sns.load_dataset('tips')

sns.barplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=df)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=df, style='smoker')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='size', data=df, style='size')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', size='size', data=df, sizes=(15, 200), alpha=0.3)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', col='time', row='smoker', data=df)
plt.show()
```
