```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='gulim')  # malgun
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 5, 16, 23, 29]

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-10, 11, 1)
y = x**2
y1 = 3*x + 50
y2 = -x + 40

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y, x, y1, x, y2)
plt.show()
```

```python
plt.figure(1, figsize=(10, 5))
plt.plot(x, y)
plt.figure(2, figsize=(10, 5))
plt.plot(x, y1)
plt.figure(1)
plt.plot(x, y2)

plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('subplot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y1)
plt.title('subplot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y2)
plt.title('subplot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.title('subplot 4')

plt.tight_layout()
plt.show()
```

```python
plt.plot(x, y)
plt.xlim(-3, 3)
plt.ylim(-20, 20)
plt.show()
```

```python
plt.plot(x, y, 'o--b', x, y1, '+-r', x, y2, 's:k', linewidth=2, markersize=7)
plt.ylim(-5, 140)
plt.title('제목')
plt.xlabel('x축')
plt.ylabel('y축')
plt.grid(True)
plt.legend(['1번', '2번', '3번'], loc='upper right')
plt.show()
```
