### 그룹

```python
df = pd.DataFrame({
    "공항": ["김포", "인천", "인천", "김포", "제주", "제주", "김해", "인천"],
    "연도": ["2020", "2020", "2019", "2018", "2020", "2019", "2020", "2020"],
    "이용객": [1444483, 182523, 214252, 244448, 1785615, 278561, 554583, 224252],
    "운항편수": [483, 623, 680, 428, 515, 571, 383, 452]
    })
df
```

```python
sg1 = df['이용객'].groupby(df['공항'])
display(sg1)
display(sg1.size(), sg1.sum(), sg1.mean())
```

```python
s = sg1.sum()
display(s, type(s))
```

```python
df1 = df.groupby(df['공항']).sum()
display(df1)

s1 = df.groupby('공항').get_group('김포').sum()
display(s1)
```

```python
sg1 = df['이용객'].groupby([df['공항'], df['연도']])
s1 = sg1.mean()

s1
```

```python
df1 = df.groupby(['공항', '연도']).sum()
display(df1)

df1 = df.groupby(['공항', '연도'])['이용객'].sum()
display(df1)

df1 = df.groupby(['공항', '연도'])['이용객'].sum()['인천']
display(df1)
```

```python
# df1 = df.drop(['연도'], axis=1).groupby(['공항']).agg(['mean', 'sum'])
df1 = df.drop(['연도'], axis=1).groupby(['공항']).agg([('평균', 'mean'), ('합계', 'sum')]) # 라벨링
df1
```

```python
df1 = pd.melt(df, id_vars=['공항', '연도'], value_vars=['이용객'], var_name='항목', value_name='값')
df1
```
