```python
df = pd.read_csv('datasets/신체치수.csv')
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 산포도
sns.relplot(x='wt', y='waist', data=df)
plt.show()
```

```python
# 박스 플롯
sns.boxplot(x='wt', data=df)
plt.show()
```

```python
# 결측치
df[['wt', 'ht', 'waist']].isnull()
```

```python
df[['wt', 'ht', 'waist']].isnull().sum()
```

```python
df[['wt', 'ht', 'waist']].isnull().sum() / len(df[['wt', 'ht', 'waist']])
```

```python
df[df['waist'].isnull()==True]
```

```python
df1 = df.dropna(subset=['waist'])
df1
```

```python
df1[df1['waist'].isnull()==True]
```

```python
df2 = df1[['wt', 'ht', 'waist']].copy()
df2
```

```python
df2.corr()
```

```python
sns.heatmap(df2.corr())
plt.show()
```

```python
sns.lmplot(x='waist', y='wt', data=df2)
plt.show()
```

```python
sns.lmplot(x='waist', y='wt', data=df2)
plt.show()
```
