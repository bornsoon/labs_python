```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[10, 20, 30, 100], [40, 50, 60, 110], [70, 80, 90, 120]],
                  columns = list('ABCD'),
                  index=pd.date_range(start='2024-01-01', periods=3))
df
```

```python
df.info()
```

```python
# 추가
df['E'] = 200
df
```

```python
df['F'] = [310, 320, 330]
df
```

```python
df['G'] = df['A'] + df['B']
df
```

```python
import datetime as dt
# to_datetime으로 DatetimeIndex로 안되고 그냥 문자열 Index가 되어버리기 때문에 converting 필요!
df.loc[pd.to_datetime('2024-01-04')] = [410, 420, 430, 440, 450, 460, 470]
df
```

```python
df.info()
```

```python
df.loc[pd.to_datetime('2024-01-05')] = [n*2 for n in df.loc['2024-01-01']]
df
```

```python
df.loc[dt.datetime(2024, 1, 6)] = [510, 520, 530, 540, 550, 560, 570]
df
```

```python
# 삭제
df1 = df.drop(['A'], axis=1)  # 인덱스 방향
df1
```

```python
df1 = df.drop(pd.to_datetime(['2024-01-01']), axis=0)  # 행 인덱스 방향
df1
```

```python
df1 = df.drop(columns = ['A', 'C'])
df1
```

```python
df1 = df.drop(index = pd.to_datetime(['20240101', '20240105']))
df1
```
