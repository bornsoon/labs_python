```python
def convert_int(el):
    result = el.replace(',', '')
    return int(result)
```

```python
df = pd.read_csv('./datasets/지점별매출.csv', index_col='일자',
                 dtype={'D지점':'int'},
                 converters={
                     'A지점':convert_int,
                     'B지점':convert_int,
                     'C지점':lambda x: int(x.replace(',', '')),
                     '운영': lambda x: np.where(x == '직영', 1, 0)
                 })
df
```

```python
# 지점별통계, 일별통계
df.loc['지점별총계'] = df.sum()
df['일별통계'] = df.sum(axis=1)
df
```
