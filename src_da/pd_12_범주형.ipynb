```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/seaborn-data/titanic.csv')
df
```

```python
# 범주 갯수 확인
df['class'].value_counts()
```

```python
df_class, df_class1 = df['class'].copy(), df[['class']].copy()
type(df_class), type(df_class1)
```

```python
df_class = df[['class']].copy()
df_class.head()
```

```python
# label encoding
# type: 순서가 있는 인코딩
df_class['cate'] = pd.Categorical(df_class['class'])
df_class['cate_code'] = df_class['cate'].cat.codes
df_class
```

```python
df_class['cate'].cat.categories, df_class['cate_code'].value_counts()
```

```python
df_class['cate_code2'] = df_class['class'].astype('category').cat.codes
df_class
```

```python
# one-hot encoding
df_class1 = pd.get_dummies(df_class, columns=['class'])
df_class1
```

```python
# label encoding
# 출현 순서로 정렬
codes, uniques = pd.factorize(df_class['class'])
codes, uniques
```

```python
df_class['class_code'] = codes
df_class
```
