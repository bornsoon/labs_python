```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='gulim')
plt.rc('figure', figsize=(3, 3))
```

```python
# 히스토그램 (개수)
s = pd.Series(np.random.randint(50, 100, size=20))

plt.subplot(1, 2, 1)
plt.hist(s)
plt.subplot(1, 2, 2)
plt.hist(s, np.arange(50, 101, 10), color='green')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 막대그래프 (값)
st = [f'{n}번' for n in range(1, 21)]
sc = pd.Series(np.random.randint(50, 100, size=20))

plt.figure(figsize=(7, 3))
plt.bar(sc.index, sc)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.barh(sc.index, sc, color=['b', 'r']*10, tick_label=st)
plt.show()
```

```python
sc2 = pd.Series(np.random.randint(50, 100, size=20))

plt.figure(figsize=(10, 5))
plt.ylim(0, 120)
plt.bar(sc.index, sc, color='y', width=0.45, align='edge', label='중간고사')
plt.bar(sc.index+0.45, sc2, color='g', width=0.45, align='edge', tick_label=st, label='기말고사')
plt.legend(loc='upper left')
plt.show()
```

```python
# 파이그래프 (동일 범주)
loc = ['강원도', '경기도', '경상도', '전라도', '제주도', '충청도']
sales = [200, 1000, 750, 500, 100, 300]

plt.pie(sales, labels=loc, autopct='%0.1f%%', startangle=90, explode=[0, 0.1, 0, 0, 0, 0], shadow=True)
plt.show()
```

```python
# 산점도 그래프 (다른 범주와 비교)

loc = ['강원도', '경기도', '경상도', '전라도', '제주도', '충청도']
sales = [200, 1000, 750, 500, 100, 300]
pops = [150, 2000, 620, 370, 67, 400]

# plt.scatter(sales, pops, s=500, c=['b', 'c', 'g', 'r', 'm', 'r'])
plt.scatter(sales, pops, s=np.array(sales)*5, c=['b', 'c', 'g', 'r', 'm', 'y'])
plt.xlabel('매출액')
plt.ylabel('인구수')

for x, y, loc in zip(sales, pops, loc):
    plt.text(x, y, loc, fontdict={'color':'yellow'})

plt.show()
```
